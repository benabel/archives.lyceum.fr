```python
def racines_du_trinome(a, b, c):
    """Renvoie les ventuelles racines d'un trinme
    
    ax**2 + bx + c
    
    Parameters
    ----------
    a : float
    b : float
    c : float
    
    Returns
    -------
    tuple
        tuple contenant les 0, 1 ou 2 racines du trinme suivant le cas
    """
    import math
    # Calcul du discriminant
    delta = b * b - 4 * a * c
    # Affichage des solutions
    if delta < 0: 
        return ()
    elif delta == 0:
        return (- b / (2 * a))
    else:
        r1 = (- b - math.sqrt(delta)) / (2 * a)
        r2 = (- b + math.sqrt(delta)) / (2 * a)
        return (r1, r2)
help(racines_du_trinome)
```

```python
def run_tests():
    # deux racines relles
    assert racines_du_trinome(1,0,-1) == (-1.0, 1.0)
    # une racine double
    assert racines_du_trinome(1, 0, 0) == (0.0)
    # pas de racine
    assert racines_du_trinome(1, 0, 1) == ()
    
    # si une assertion s'avre fausse la fonction va renvoyer une AssertionError
    print("Tous les tests ont t pass avec succs")
run_tests()
```
